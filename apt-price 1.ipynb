```python
# data/apt-price.xlsx 불러오기
import pandas as pd

df = pd.read_excel('data/apt-price.xlsx', header=12)
df.head()
```

```python
df.info()
```

```python
# df에서 'No'컬럼을 index로 설정
df.set_index('NO', inplace=True)
df.head()
```

```python
df.tail()
```

```python
# df에서 '시군구'컬럼 삭제
df.drop(columns=['시군구'], inplace=True)
df.head()
```

```python
df.drop(columns=['번지'], inplace=True)
df.head()
```

```python
# df['단지명']중에 '파크리오'만 선택
# boolean indexing
df2 = df[df['단지명'] == '파크리오']
df2.head()
```

```python
# 기존 인덱스를 삭제(drop)하고, 인덱스를 재설정
df2.reset_index(drop=True, inplace=True)
df2.head()
```

```python
df2.drop(columns=['건축년도','도로명','주택유형'], inplace=True)
df2.head()
```

```python
# df2['해제사유발생일']이 '-' 아닌 행을 찾기
df2[df2['해제사유발생일'] != '-']
```

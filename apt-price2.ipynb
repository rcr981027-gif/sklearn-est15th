```python
# data/apt-price.xlsx 불러오기
import pandas as pd

df = pd.read_excel('data/apt-price.xlsx', header=12)
df.head()
```

```python
df.info()
```

```python
# df에서 'No'컬럼을 index로 설정
df.set_index('NO', inplace=True)
df.head()
```

```python
df.tail()
```

```python
# df['단지명']중에 '파크리오'만 선택
# boolean indexing
df2 = df[df['단지명'] == '파크리오']
df2.head()
```

```python
# df2에서 '시군구','번지','매수자','매도자','건축년도','도로명','주택유형' 컬럼 삭제 후 df2에 저장
df2 = df2.drop(['시군구','본번','부번','번지','매수자','매도자','건축년도','도로명','주택유형'], axis=1)
df2.head()
```

```python
# df2['해제사유발생일']!='-' 조건의 행 삭제
df3 = df2[df2['해제사유발생일']!='-']
df3.head()
```

```python
# 기존 인덱스를 삭제(drop)하고, 인덱스를 재설정
df3.reset_index(drop=True, inplace=True)
df3.head()
```

```python
# df3['전용면적(㎡)']의 컬럼명을 '전용면적'으로 수정
df3.rename(columns={'전용면적(㎡)': '전용면적'}, inplace=True)
df3.head()
```

```python
# #df3의 '계약년월'과 '계약일'을 하나의 문자열로 합치고, datetime으로 변환
df3['계약년월일'] = pd.to_datetime(df3['계약년월'].astype(str) + df3['계약일'].astype(str))
df3.head()
```

```python
# matplotlib에서 컴퓨터에 설치된 폰트 보기
import matplotlib
matplotlib.font_manager.findSystemFonts(fontpaths=None, fontext='ttf')
```

```python
import matplotlib.pyplot as plt
# 폰트 설정
plt.rc('font', family='Malgun Gothic')
plt.rc('axes', unicode_minus=False)
# 1. 데이터 전처리 (안전장치)
# 거래금액 쉼표 제거 및 정수 변환
df3['거래금액(만원)'] = df3['거래금액(만원)'].astype(str).str.replace(',', '').astype(int)
plt.figure(figsize=(12, 6))
# 2. 산점도 그리기
# c: 색상을 결정할 컬럼 (전용면적)
# cmap: 색상 테마 (예: 'viridis', 'plasma', 'coolwarm' 등)
# alpha: 투명도 (점들이 겹칠 때 구분을 위함)
scatter = plt.scatter(
    df3['계약년월일'], 
    df3['거래금액(만원)'], 
    c=df3['전용면적'], 
    cmap='viridis', 
    alpha=0.7
)
plt.xticks(rotation=90)
plt.xlabel('날짜')
plt.ylabel('거래금액(만원)')
plt.grid(True, alpha=0.3)
# 3. 컬러바(Legend 역할) 추가
cbar = plt.colorbar(scatter)
cbar.set_label('전용면적(㎡)') # 컬러바 옆에 라벨 붙이기
```
